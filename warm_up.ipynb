## Problem 1 - Loading and Displaying a Toy Problem

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
import glob
```

```python
files = sorted(glob.glob("toy_problem/*.png"))
frames = []

for f in files:
    img = io.imread(f) # load
    img = color.rgb2gray(img) # convert to grayscale
    img = img.astype(np.float32) # float
    frames.append(img)

V = np.stack(frames, axis=2) # (256, 256, 64)

# displaying frames
plt.figure()
for t in range(10): 
    plt.clf()
    plt.imshow(V[:, :, t], cmap='gray')
    plt.pause(0.05)

plt.show()
```

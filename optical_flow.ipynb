### Part 2 - Image Gradients

```python
import PIL 
from PIL import Image
import io
import numpy as np
```

#### Problem 2.1 - Low Level Gradient Calculation

```python
images = []
for i in range(1,65):
    if i < 10:
        image_path = f"toy_problem/frame_0{i}.png"
    else:
        image_path = f"toy_problem/frame_{i}.png"
    image1 = PIL.Image.open(image_path)
    img_bl = image1.convert("L")
    image_array = np.array(img_bl)/255
    images.append(image_array)

```

```python
for image in images:
    dx = image[1:, :] - image[:-1, :]
    dy = image[:, 1:] - image[:, :-1]
    print(dx,dy)
```
